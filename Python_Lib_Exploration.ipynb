Exploring mathplotlib and plotly

```python
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px

# Load the sample dataset
iris = sns.load_dataset("iris")
```

#mathplotlib

```python
plt.figure()
scatter = plt.scatter(
    x=iris['sepal_length'],
    y=iris['sepal_width'],
    c=iris['species'].map({"setosa": "r", "versicolor": "g", "virginica": "b"}),
    alpha=0.7
)

plt.title("Sepal Length vs Sepal Width")
plt.xlabel("Length")
plt.ylabel("Width")
plt.show()
```

# plotly

```python
fig = px.scatter(
    iris, x="sepal_length", y="sepal_width", color="species",
    title="Sepal Length vs Sepal Width",
    labels={"sepal_length": "Sepal Length", "sepal_width": "Sepal Width"},
)
fig.show()
```
